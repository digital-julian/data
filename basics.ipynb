```python
FILEPATH = "data/task_data/2021-11-07.json"
import json

def get_dataframe_from_json(json_dict):
    from datetime  import datetime 
    import pandas
    dataframe = pandas.json_normalize(json_dict)
    dataframe.index =  pandas.to_datetime(dataframe['date'])

    # drop unix timestamps & just use ISO8601 datetimes
    dataframe = dataframe.loc['1999-01-01':]
    return dataframe

with open(FILEPATH) as json_file:
    data = json.load(json_file)
```

```python
# print out graph of exp over time
exp_df = get_dataframe_from_json(data['history']['exp'])
print("rough estimate of app usage over time:")
display(exp_df.plot())
```

```python
exp_df = get_dataframe_from_json(data['history']['todos'])
print("rough estimate of todo procrastination:")
display(exp_df.plot())
```
